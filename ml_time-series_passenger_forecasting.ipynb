The initial autocorrelation test shows that at around 15 Lag, the correlation value, $R$ value, falls within the confidence threshold. ### Preliminary ARIMA calculations 

The $d$ parameter is the integration "differences" parameter. ARIMA develops on this with an integration filter, which establishes this parameter $d$. Technically $d$, is the number of times differences have been taken between current and previous values. The relevance of $d$ is that it allows us to effectively remove any trend or seasonal components that might be present in the data, making it stationary. This appears in the order parameter of `.ARIMA` method as `order=(p,d,q)`.

```python
# ADF Test:
diff_test_adf=ndiffs(passengers, test='adf')
diff_test_kpss=ndiffs(passengers, test='kpss')
print("d for adf: ", diff_test_adf,"\nd for kpss:", diff_test_kpss)
```

```python
mod = sm.tsa.arima.ARIMA(endog=passengers, order=(1, 1, 0), freq="M") 
res = mod.fit()
print(res.summary())
```

```python
mod = STLForecast(
  passengers, sm.tsa.arima.ARIMA,
  model_kwargs=dict(order=(1, 1, 10), trend="t",freq="M")
)
res = mod.fit().model_result
print(res.summary())
```

```python
STEPS = 20
forecasts_df = res.get_forecast(steps=STEPS).summary_frame() 
```

```python
ax = passengers.plot(figsize=(12, 6))
plt.ylabel('SPY')
forecasts_df['mean'].plot(style='k--')
ax.fill_between(
    forecasts_df.index,
    forecasts_df['mean_ci_lower'],
    forecasts_df['mean_ci_upper'],
    color='k',
    alpha=0.1
)
```

To understand the best parameter choices for (p,q), having seen from the ndiffs tests that $d$ should be either 0 or 1, we could coarsely scan the available options and create a forecast. As a measure to check, we would look at the Akaike information criterion (AIC), after inserting the tuple. The lower the better. In an example below we scan over the $q$ index, setting $p=2$ throughout. 

```python
forecasts = []
qs = []
aic_scores = []
for q in range(0, 10, 1):
    mod = STLForecast(
            passengers, sm.tsa.arima.ARIMA, 
            model_kwargs=dict(order=(1, 1, q), trend="t",freq="M")
        )
    res = mod.fit().model_result
    print(f"aic ({q}): {res.aic}")
    aic_scores.append(res.aic)
    forecasts.append(
            res.get_forecast(steps=STEPS).summary_frame()
        )
    qs.append(q)
forecasts_df = pd.concat(forecasts, axis=1)
forecasts_df.columns = qs
```

```python
# Model definition

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Input(len(passengers.columns))) #Using len(passengers.columns) means that if we feature engineer to make more features, we won't have to change it.
model.add(tf.keras.layers.Dropout(DROPOUT_RATIO))
model.add(tf.keras.layers.Dense(HIDDEN_NEURONS)) #Note no relu here, very simple first trial
model.add(tf.keras.layers.Dropout(DROPOUT_RATIO))            
model.add(tf.keras.layers.Dense(1))

loss_func = tf.keras.losses.mean_squared_error
opt = tf.keras.optimizers.Adam()

model.compile(loss=loss_func, optimizer=opt, metrics=[keras.metrics.RootMeanSquaredError(), keras.metrics.MeanAbsoluteError()])
```

## 4/. Splitting features

Let us make month and year features for our dataframe.

```python
# Data import
passengers = pd.read_csv(
  "passengers.csv", parse_dates=["date"]
).set_index("date")
```

```python
# Now the dataframe has three features
passengers["month"] = passengers.index.month.values
passengers["year"] = passengers.index.year.values
passengers.head()
```

Having set the input of the network based on the number of columns in the dataframe, we do not have to change anything, in terms of input shape.

```python
# Model definition and parameters 

DROPOUT_RATIO = 0.2
HIDDEN_NEURONS = 20 # The number of nodes in the hidden layer. Start small - remmeber the dataset is not large: 1 feature and small dataset shouldn't need much complexity
#lr = 0.0001
epoch_num = 100
callback = tf.keras.callbacks.EarlyStopping(
  monitor='loss', patience=3
) # early stopping means if we see no improvement in the loss after 3 epochs, we terminate further optimisation and stop there.


def create_model(passengers):
    model = tf.keras.models.Sequential()
    model.add(tf.keras.layers.Input(len(passengers.columns))) #Using len(passengers.columns) means that if we feature engineer to make more features, we won't have to change it.
    model.add(tf.keras.layers.Dropout(DROPOUT_RATIO))
    model.add(tf.keras.layers.Dense(HIDDEN_NEURONS, activation='relu'))
    model.add(tf.keras.layers.Dropout(DROPOUT_RATIO))            
    model.add(tf.keras.layers.Dense(1))
    
    loss_func = tf.keras.losses.mean_squared_error
    opt = tf.keras.optimizers.Adagrad()
    
    model.compile(loss=loss_func, optimizer=opt, metrics=[keras.metrics.RootMeanSquaredError(), keras.metrics.MeanAbsoluteError()])
    return model
```

```python
# Generating a model from our definition
model = create_model(passengers)

# We also generate new X_train, X_test etc. since we have 3 features present now.
X_train, X_test, y_train, y_test = train_test_split(passengers, passengers.passengers.shift(-1), shuffle=False)
X_train.head()
X_test.head()
```

```python
history = model.fit(x=X_train, y=y_train, epochs=epoch_num, callbacks=[callback])
```

```python
pred = model.predict(X_test)

def show_result(y_test, predicted):
  plt.figure(figsize=(13, 6.5))
  plt.plot(y_test.index, predicted, 'o-', label="predicted")
  plt.plot(y_test.index, y_test, '.-', label="actual")
  plt.ylabel("Passengers")
  plt.legend()


show_result(y_test,pred)
```

## 5/. Further improvements: embedding, scaling

The baseline is marginally better, but not much. We can do some further work to see how far we can improve this. Changes include

- Embedding of the month feature.
- Treat year as a linear predictor.
- Add previous month's passengers to our predictions.
- Scale our predictions based on the standard deviation in the training dataset.

**Embedding of the month feature**: We know that there are lots of travelers in both December and January, but perhaps a much lower volume in November. We can capture these relationships based on the data. This can be done in an embedding layer. An embedding layer is a mapping of distinct categories to real numbers. This mapping is updated as part of the network optimization.

**Treat year as a linear predictor**: The year is closely related to the overall trend. Each year, the number of airline passengers increases. We can model this relationship non-linearly or linearly. Here, I've decided to just model a linear relationship between the year feature and the outcome.

**Add previous month's passengers to our predictions**: The relationship between the previous month's passenger numbers and those in this month is again assumed to be linear.

**Scale our predictions**: Finally, we can scale our predictions, similar to the inverse transformation of the standard transformation. Standard transformation $z = (x - \mu)/\sigma$. Our formula will be  $\hat{v_{t}} = p \sigma + v_{t-1}$ where $\hat{v_{t}}$ are the airline passengers at time $t$ and $p$ is the predicition based on the embedded month and the year.

```python
# Parameters 
DROPOUT_RATIO = 0.2
HIDDEN_NEURONS = 10 # The number of nodes in the hidden layer. Start small - remmeber the dataset is not large: 1 feature and small dataset shouldn't need much complexity
#lr = 0.0001
epoch_num = 100
callback = tf.keras.callbacks.EarlyStopping(
  monitor='loss', patience=3
) # early stopping means if we see no improvement in the loss after 3 epochs, we terminate further optimisation and stop there.
```

```python
# Data import
passengers = pd.read_csv(
  "passengers.csv", parse_dates=["date"]
).set_index("date")

# Now the dataframe has three features
passengers["month"] = passengers.index.month.values
passengers["year"] = passengers.index.year.values
passengers.head()
```

```python
tf.constant(passengers.passengers.std())
```

Let us redefine the network based on the feature changes we are going to make.

- We are passing months through an embedding layer, so we make that a separate input.
- The year is also passed first through a single neuron hence the Dense(1) definition.
- The cont_layer is added back in later as mentioned above.

**Note**: sequential does not handle multiple inputs (as far as I am aware), therefore we define the network slightly differently.

```python
def create_model_v2(train):
  scale = tf.constant(train.passengers.std(), dtype=np.float32) # Need to define the data type here to float32, was conflicting with float64
  cont_layer = Input(shape=1)
    
  cat_layer = Input(shape=1) #month to reals input
  embedded = Embedding(13, 5)(cat_layer) # cat_layer passes through the embedding layer the 12 months into real numbers
  emb_flat = Flatten()(embedded) #flattening month layer
    
  year_input = Input(shape=1)
  year_layer = Dense(1)(year_input) # this is the linear projection, we take the year_input and form w_{year}*year_input + b_{year} at the node
    
  hidden_output = Concatenate(-1)([emb_flat, year_layer, cont_layer]) # Joining our inputs together
  output_layer = keras.layers.Dense(1)(hidden_output)
  output = output_layer * scale + cont_layer # output_layer is like the t time prediction, while cont_layer is the t-1 time prediction
  model = keras.models.Model(inputs=[cont_layer, cat_layer, year_input], outputs=output)
    
  model.compile(loss='mse', optimizer=keras.optimizers.Adam(), metrics=[RootMeanSquaredError(), MeanAbsoluteError()])
  return model
```

```python
# We also generate new X_train, X_test etc. since we have 3 features present now.
X_train, X_test, y_train, y_test = train_test_split(passengers, passengers.passengers.shift(-1), shuffle=False)
X_train.head()
X_test.head()
```

```python
model= create_model_v2(X_train)
```

```python
history = model.fit((X_train["passengers"], X_train["month"], X_train["year"]), y_train, epochs=4000, callbacks=[callback],)
```

```python
plt.title("Loss history")
plt.xlabel("Epoch number")
plt.ylabel("Log(Loss)")
plt.plot(np.log(history.history["loss"]))
```

```python
pred = model.predict((X_test["passengers"], X_test["month"], X_test["year"]))
```

```python
def show_result(y_test, predicted):
  plt.figure(figsize=(13, 6.5))
  plt.plot(y_test.index, predicted, 'o-', label="predicted")
  plt.plot(y_test.index, y_test, '.-', label="actual")
  plt.ylabel("Passengers")
  plt.legend()


show_result(y_test,pred)
```

So here we see that the output is far better capturing baseline information as well as seasonal info.
